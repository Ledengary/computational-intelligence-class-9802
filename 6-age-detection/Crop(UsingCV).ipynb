```python
import cv2
import numpy as np
import os
from tqdm import tqdm
from PIL import Image
# import matplotlib as plt
import matplotlib.pyplot as plt 

DATADIR = "/home/omid/Ex jupyter notebook/machine learning/train"
DATADIR2 = "/home/omid/Ex jupyter notebook/machine learning/final"

IMG_SIZE = 150
for category in range(1, 7): 
    path = os.path.join(DATADIR,('0'+ str(category)))
    path2 = os.path.join(DATADIR2,('0'+ str(category)))

    for img in tqdm(os.listdir(path)):
        try:
            image = cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE )  # convert to array
            new_array = cv2.resize(image, (IMG_SIZE, IMG_SIZE))  # resize to normalize data size
            # Convert the image to RGB colorspace
            image = cv2.cvtColor(new_array, cv2.IMREAD_GRAYSCALE)
            face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
            # Convert the image to gray 
            gray_image = cv2.cvtColor(image, cv2.IMREAD_GRAYSCALE)
            # Detect faces in the image using pre-trained face dectector
            faces = face_cascade.detectMultiScale(gray_image, 1.25, 6)

            # Get the bounding box for each detected face
            face_crop= None
            if len(faces) > 0 :
                for f in faces:
                    x, y, w, h = [ v for v in f ]
                    cv2.rectangle(gray_image, (x,y), (x+w, y+h), (255,0,0), 3) 
                    face_crop = gray_image[y:y+h, x:x+w]

                im = Image.fromarray(face_crop)
                jp =img.split('.')
                im.save(os.path.join(path2, (jp[0]+'.png')))
            
        except Exception as e: 
            print('hi')
            print(e)
```
